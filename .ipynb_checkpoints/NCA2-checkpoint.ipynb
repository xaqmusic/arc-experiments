# 3D NCA Cube Experiments

## Load Data

```python
import json
import numpy as np
from typing import Dict, List
import matplotlib.pyplot as plt

def load_data(json_path: str) -> Dict:
    """Load data from JSON file containing input/output grid pairs"""
    
    # Load entire JSON data
    with open(json_path, 'r') as f:
        data = json.load(f)
    
    train_examples: List[Dict] = []
    
    # Process all training examples (assuming they're top-level objects)
    for entry in data:
        if isinstance(entry, dict):
            input_grid = np.array(entry['input'])
            output_grid = np.array(entry['output'])
            
            # Calculate ratio between inputs and outputs
            try:
                ratio = np.mean(input_grid / output_grid)
            except ZeroDivisionError:
                ratio = float('nan')  # Handle division by zero
            
            train_examples.append({
                'input': input_grid,
                'output': output_grid,
                'ratio': ratio, 
                'steps': 0  # Initialize steps for training
            })
    
    return {
        'train_examples': train_examples,
    }

block1 = load_data("/home/xaqmusic/ARC-AGI/data/training/00d62c1b.json")
```

## Cell Definition

```python
class NCACell:
    def __init__(self, num_input_channels: int, hidden_size: int):
        self.num_input_channels = num_input_channels
        self.hidden_size = hidden_size
        
        # Neural network layers
        self.fc1 = np.random.randn(num_input_channels, hidden_size)
        self.fc2 = np.random.randn(hidden_size, 1)  # Output layer
        
        # Learning rate
        self.learning_rate = 0.01

    def forward(self, input: np.ndarray) -> np.ndarray:
        """Process current state and return new state"""
        # Combine input with its neighborhood (for now just input itself)
        combined_input = np.reshape(input, (-1,))
        
        # Simple neural network processing
        z = np.dot(combined_input, self.fc1)
        a = np.maximum(z, 0)  # ReLU activation
        
        output = np.dot(a, self.fc2)
        return np.clip(output, 0, 9)  # Ensure values stay between 0 and 9

    def get_current_state(self) -> np.ndarray:
        """Return current state of the cell"""
        return np.clip(np.dot(self.get_activation(), self.fc1), 0, 9)

    def get_age(self) -> int:
        """Return how many steps the cell has been alive"""
        return self.steps

    def update(self, loss: float):
        """Update weights based on computed loss"""
        pass  # Weight updates will be implemented later

block2 = NCACell(num_input_channels=1, hidden_size=16)
```

## Grid Init

```python
import json
import numpy as np
from typing import Dict, List
import matplotlib.pyplot as plt

def load_data(json_path: str) -> Dict:
    """Load data from JSON file containing input/output grid pairs"""
    with open(json_path, 'r') as f:
        data = json.load(f)
    
    return {
        'train_examples': [entry['input'] for entry in data if isinstance(entry, dict)],
        # Handle test example separately
    }

block1 = load_data("/home/xaqmusic/ARC-AGI/data/training/00d62c1b.json")

# Now let's update block 3 with proper grid initialization and visualization

def initialize_nca_grid(input_grid: np.ndarray,
                       train_examples: List[np.ndarray]) -> Dict:
    """Initialize the NCA grid with training examples"""
    
    # Get dimensions based on input grid
    rows = len(input_grid)
    cols = len(input_grid[0])
    
    # Create a 3D cube filled with random values between 0 and 9
    nca_grid = np.random.rand(rows, cols, 1)  # [height, width, depth]
    
    # Set up training examples: Face A is input face (training example)
    for i in range(len(train_examples)):
        current_input = train_examples[i]
        
        # Create a new grid with this as the front layer
        nca_grid_layer = np.random.rand(rows, cols, 1)  # Random initialization
        
        # Set front face to input grid
        if len(current_input.shape) == 2:  # If input is 2D, reshape it for consistency
            current_input_3d = current_input.reshape(-1, 1)
        else:
            current_input_3d = current_input
        
        nca_grid_layer[0] = current_input_3d
        
        # Update NCA grid for next iteration
        nca_grid = np.concatenate([nca_grid_layer, nca_grid], axis=2)  # Combine new layer with old layers
        nca_grid = nca_grid[-4:]  # Keep only the last few layers
    
    return {
        'nca_grid': nca_grid,
        'current_step': 0
    }

# Initialize NCA grid and print shapes for debugging
block3 = initialize_nca_grid(block1['input'], block1['train_examples'])
for i, input_grid in enumerate(block1['train_examples']):
    print(f"Training example {i+1} has shape:", input_grid.shape)
    # Visualize each face of the cube at initialization step 0
    plt.figure(figsize=(5,5))
    plt.imshow(input_grid, cmap='hot')
    plt.colorbar()
    plt.title(f"Input Face {i+1}")
    plt.show()

plt.figure(figsize=(5,5))
plt.imshow(block3['nca_grid'][0], cmap='hot')  # Output face
plt.colorbar()
plt.title("Output Face")
plt.show()

```

## Training Loop

```python
def train_nca(nca_grid: np.ndarray, 
              train_examples: List[Dict], 
              test_example: Dict, 
              num_epochs: int = 100) -> Dict:
    """Main training loop for the NCA"""
    
    # Initialize loss history
    loss_history = {'train': [], 'test': []}
    
    for epoch in range(num_epochs):
        current_step = nca_grid['current_step']
        
        # Evolve the NCA for one step
        new_nca_grid = evolve_step(nca_grid)
        nca_grid['nca_grid'] = new_nca_grid
        nca_grid['current_step'] += 1
        
        # Visualize if it's the first or every 10th step
        if epoch % 10 == 0:
            visualize_nca(nca_grid, epoch)
        
        # Check loss on test example after each epoch
        train_loss = compute_loss(nca_grid['nca_grid'], train_examples[0]['input'],
                                  train_examples[0]['output'])
        test_loss = compute_loss(nca_grid['nca_grid'], test_example['input'],
                                 test_example['output'])
        
        loss_history['train'].append(train_loss)
        loss_history['test'].append(test_loss)
        
    return nca_grid, loss_history

def evolve_step(nca_grid: np.ndarray) -> np.ndarray:
    """Run one step of NCA evolution"""
    
    # Update each cell based on its neighborhood
    rows = nca_grid.shape[0]
    cols = nca_grid.shape[1]
    
    new_grid = np.zeros_like(nca_grid)
    
    for i in range(rows):
        for j in range(cols):
            # Get Moore neighborhood (including diagonals)
            neighborhood = get_neighborhood(nca_grid, i, j)
            
            # Combine current state and neighborhood
            input_to_network = np.concatenate((neighborhood.flatten(), [current_age(i, j)]))
            
            # Process through neural network to get new state
            new_state = block2.forward(input_to_network)
            
            new_grid[i, j] = new_state[0][0]
    
    return new_grid

def current_age(i: int, j: int) -> int:
    """Return the age of a cell (number of steps it has been alive)"""
    return block3['current_step']

def get_neighborhood(grid: np.ndarray, i: int, j: int) -> np.ndarray:
    """Extract Moore neighborhood from grid for cell at (i,j)"""
    
    rows = grid.shape[0]
    cols = grid.shape[1]
    
    # Create a mask of indices in the neighborhood
    y_coords = np.arange(max(0, i-1), min(rows, i+2))
    x_coords = np.arange(max(0, j-1), min(cols, j+2))
    idxs = np.meshgrid(y_coords, x_coords)
    
    # Get values from these indices, flattened
    neighborhood = grid[idxs[0], idxs[1]].flatten()
    
    return neighborhood

def compute_loss(predicted_grid: np.ndarray, 
                 actual_grid: np.ndarray) -> float:
    """Compute mean squared error between predicted and actual grids"""
    
    return np.mean((predicted_grid - actual_grid)**2)

block4 = {
    'nca_grid': block3['nca_grid'],
    'train_examples': block1['train_examples'],
    'test_example': block1['test_example'],
    'num_epochs': 100
}
block4, loss_history = train_nca(block4['nca_grid'], 
                              block4['train_examples'],
                              block4['test_example'],
                              num_epochs=block4['num_epochs'])
```

## Save State

```python
def save_state(nca_grid: Dict, model: Dict, epoch: int) -> None:
    """Save NCA state and trained model to JSON"""
    
    # Save cube state
    output = f'nca_epoch_{epoch}.json'
    with open(output, 'w') as f:
        json.dump(nca_grid, f)
        
    # Save model weights
    model_path = 'model.pth'
    np.save(model_path, block2.fc1)  # Assuming weights are in fc1 layer

block5 = save_state(block4['nca_grid'], block2, 0)
```

## convert to numpy arrays

```python
def convert_to_numpy(data: Dict) -> Dict:
    """Convert all values in a dictionary to NumPy arrays or scalars"""
    converted = {}
    
    for key, value in data.items():
        try:
            if isinstance(value, np.ndarray):
                # If it's already a numpy array, ensure it's not empty
                converted[key] = value if value.size else []
            elif isinstance(value, list):
                # Convert lists to numpy arrays recursively but ensure shape consistency
                sub_list_shapes = []
                is_list_like = True
                for item in value:
                    try:
                        if isinstance(item, (list, np.ndarray)):
                            if isinstance(item, list):
                                converted_item = convert_to_numpy([item])[0]
                            else:
                                converted_item = item
                            sub_list_shapes.append(converted_item.shape)
                        else:
                            converted_item = np.array(item)
                    except Exception as e:
                        print(f"Error converting non-list/non-ndarray: {item} at position {value.index(item)} with type {type(item)}")
                        is_list_like = False
                        break
                    if isinstance(converted_item, np.ndarray):
                        # Check if we have a list of arrays and they are consistent
                        if first_array is None:
                            first_array = converted_item
                        elif not np.array_equal(converted_item, first_array):
                            print(f"Inhomogeneous shapes detected in {key}. Expected shape: {first_array.shape}, Actual shape: {converted_item.shape}")
                            raise ValueError("Inconsistent array lengths")
                # If we have a list of arrays and they are consistent, replace the value
                if is_list_like:
                    converted[key] = [first_array]
            else:
                # Handle non-array types (e.g., integers, strings)
                converted[key] = np.array(value)
        except Exception as e:
            print(f"Error converting {key}: {str(e)}")
            raise
    
    return converted

# Convert block1 to NumPy arrays
block1 = convert_to_numpy(block1)

# Verify the conversion worked and inspect shapes
for key, value in block1.items():
    if isinstance(value, np.ndarray):
        print(f"Key: {key}, Shape: {value.shape}")
    elif isinstance(value, list):
        print(f"List type - not converted to numpy array\n{value}")
    else:
        print(f"Type: {type(value)}\n")
```
