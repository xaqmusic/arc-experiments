## Data Visualizations

List all of the training data and visualize input/output grids:

```python
import os
import json
import matplotlib.pyplot as plt

# Training data directory path (ensure this is correct)
training_dir = "/home/xaqmusic/ARC-AGI/data/training"

def list_training_files(training_dir):
    """List all JSON files in the training directory"""
    try:
        json_files = [f for f in os.listdir(training_dir) if f.endswith('.json')]
        
        print(f"Found {len(json_files)} JSON files")
        return json_files
    except Exception as e:
        print(f"Error listing files: {str(e)}")
        return []

def visualize_data(data):
    try:
        plt.figure(figsize=(12, 6))
        
        # Visualize the input grid
        plt.subplot(1, 2, 1)
        plt.imshow(data['input'], cmap='viridis')
        plt.colorbar(label='Cell Values')
        plt.title('Input Grid')
        plt.tight_layout()
        
        # Visualize the output grid
        plt.subplot(1, 2, 2)
        plt.imshow(data['output'], cmap='viridis')
        plt.colorbar(label='Cell Values')
        plt.title('Output Grid') 
        plt.tight_layout()
        
        plt.show()
    except Exception as e:
        print(f"Error processing data: {str(e)}")

# Get all JSON files
json_files = list_training_files(training_dir)

if json_files:
    # Choose the first valid JSON file to open
    for json_file in json_files:
        full_path = os.path.join(training_dir, json_file)
        
        with open(full_path, 'r') as file:
            data = json.load(file)
            
            # Assuming each entry is a dictionary with keys 'input' and 'output'
            for entry in data['train'] + data['test']:
                print(f"File name: {json_file}")  # Add this line to list the file name
                visualize_data(entry)
else:
    print("No JSON files found in the training directory.")
```

## Data Conversion

load each json file in the training folder and output each input/output tensor pair into a test and training pth files

```python
import os
import json
import numpy as np
import torch

directory_path = "/home/xaqmusic/ARC-AGI/data/training"

def list_training_files(training_dir):
    """List all JSON files in the training directory"""
    try:
        json_files = [f for f in os.listdir(training_dir) if f.endswith('.json')]
        
        print(f"Found {len(json_files)} JSON files")
        return json_files
    except Exception as e:
        print(f"Error listing files: {str(e)}")
        return []

def process_training_data(directory_path):
    """Process JSON training files and extract input/output pairs for both train and test sections."""
    try:
        all_pairs = []  # List to store all extracted pairs
        
        for filename in os.listdir(directory_path):            
            # Load the JSON file
            try:
                full_path = os.path.join(directory_path, filename)
                with open(full_path, 'r') as json_file:
                    print(f"Processing {filename}")
                    data = json.load(json_file)
                    
                    # Process both train and test sections
                    for section in ['train', 'test']:
                        if section in data:  # Ensure the section exists to avoid errors
                            input_data = data[section][0]['input']
                            output_data = data[section][0]['output']
                            
                            # Convert numpy arrays to tensors
                            input_tensor = torch.tensor(input_data, dtype=torch.float32)
                            output_tensor = torch.tensor(output_data, dtype=torch.float32)
                            
                            # Create a dictionary with section identifier
                            pairs_dict = {
                                f'{section}_input': input_tensor,
                                f'{section}_output': output_tensor,
                                'filename': filename
                            }
                            
                            all_pairs.append(pairs_dict)
                    
            except json.JSONDecodeError:
                print(f"Skipping file {filename} due to invalid JSON format")
                
        # Save train and test data separately
        torch.save([pairs for pairs in all_pairs if pairs['filename'].startswith('train')], 
                   'training_data.pth')
        torch.save([pairs for pairs in all_pairs if not pairs['filename'].startswith('train')],
                   'test_data.pth')
        print("Training and test data saved successfully.")
        
    except Exception as e:
        print(f"An error occurred during processing: {str(e)}")

list_training_files(directory_path)
process_training_data(directory_path)
```

## Step by Step for loading in a single json and setting up a cube of NCA voxels

```python
import json
import numpy as np

# Load the JSON data from the file
with open("/home/xaqmusic/ARC-AGI/data/training/00d62c1b.json", "r") as f:
    data = json.load(f)

# Access the training data (input and output grids)
train_data = data["train"]
input_grid1 = train_data[0]["input"]
output_grid1 = train_data[0]["output"]

# Convert lists of lists to NumPy arrays
input_array = np.array(input_grid1)
output_array = np.array(output_grid1)

# Get the dimensions of the input and output grids
input_shape = input_array.shape
output_shape = output_array.shape

print("Input shape:", input_shape)
print("Output shape:", output_shape)

# Determine the maximum grid size for the cube faces
max_size = max(input_shape[0], input_shape[1])
print("Max Size:", max_size)

```

```python
import numpy as np

# Define the number of layers in the NCA cube
num_layers = max_size

# Create an empty voxel cube (3D array)
voxel_cube = np.zeros((max_size, max_size, num_layers))

# Initialize face A with the input grid
voxel_cube[:, :, 0] = np.rot90(input_grid1, k=-1)

```
